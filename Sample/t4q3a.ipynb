# Tutorial 4, Question 3

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import pylab as plt
from mpl_toolkits.mplot3d import Axes3D

import os
if not os.path.isdir('figures'):
    os.makedirs('figures')
```

## Set learning parameters

```python
lr = 0.01
num_epochs = 3000

seed = 100
np.random.seed(seed)
tf.random.set_seed(seed)
```

## Generate training data

```python
X = np.zeros((9*9, 2)).astype(np.float32)
p = 0
for i in np.arange(-1, 1.001, 0.25):
    for j in np.arange(-1, 1.001, 0.25):
        X[p] = [i, j]
        p += 1

np.random.shuffle(X)
Y = np.zeros((9*9, 1)).astype(np.float32)
Y[:,0] = 0.8*X[:,0]**2 - X[:,1]**3 + 2.5*X[:,0]*X[:,1]
```

## Plot training inputs

```python
plt.figure(1)
plt.plot(X[:,0], X[:,1], 'rx')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Training inputs')
plt.savefig('./figures/t4q3a_1.png')
```

## Plot targets

```python
fig = plt.figure(2)
ax = fig.gca(projection = '3d')
ax.scatter(X[:,0], X[:,1], Y[:,0], color='blue', marker='.')
ax.set_title('Targets for training')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$\phi$')
ax.set_xticks([-1.0, -0.5, 0, 0.5, 1.0])
ax.set_yticks([-1.0, -0.5, 0, 0.5, 1.0])
plt.savefig('./figures/t4q3a_2.png')
```

## Build the neural network

```python
network = Sequential([
    Dense(10, activation='relu'),
    Dense(5, activation='relu'),
    Dense(1, activation='linear')
])
```

## Set the optimizer, cost, measures, etc. for learning

```python
network.compile(optimizer='sgd',
              loss='mse')
```

## Learn the model

```python
history = network.fit(X, Y, 
                    epochs=num_epochs, 
                    verbose = 2, 
                    use_multiprocessing=True)
```

```python
history.history
```

## Plot learning curves

```python
plt.figure(3)
plt.plot(range(num_epochs), history.history['loss'])
plt.xlabel('iterations')
plt.ylabel('error (mse)')
plt.title('GD learning')
plt.savefig('figures/t4q3a_3.png')
```

```python
# Print final errors
print('train m.s.e. = %3.3f'%(history.history['loss'][-1]))
```

```python
# Find network predictions
pred = network(X)
```

## Plot targets and predicted points

```python
fig = plt.figure(4)
ax = fig.gca(projection = '3d')
ax.scatter(X[:,0], X[:,1], Y[:,0], color='blue', marker='.', label='targets')
ax.scatter(X[:,0], X[:,1], pred[:,0], color='red', marker='x', label='predictions')
ax.set_title('Targets and Predictions')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$\phi$')
ax.set_xticks([-1.0, -0.5, 0, 0.5, 1.0])
ax.set_yticks([-1.0, -0.5, 0, 0.5, 1.0])
ax.legend()
plt.savefig('./figures/t4q3a_4.png')
```
